# Exploratory Data Analysis (EDA) & Visualizations
## Project: The AI-Labor Paradox (2024–2026)

This notebook provides a detailed visual exploration of the synthesized dataset, designed to highlight the K-Shaped displacement trajectories, the Salary Paradox, and the efficacy of the Skill Shield.

### 1. Setup & Environment Initialization


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure Seaborn Aesthetics
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 6)

```

### 2. Loading the Ultimate Gist Dataset
Mapping the generated synthetic master payload into a Pandas DataFrame.


```python
# Load the dataset from the 03_final_data repository
file_path = r's:\MISSION FIELD\AI - ALGOMA\FINAL PROJECT\DATABASE\03_final_data\ai_labor_paradox_ultimate_gist.csv'
df = pd.read_csv(file_path)
display(df.head())
print(f'\nDataset Shape: {df.shape}')

```

## Task 1: Univariate Analysis (The Landscape)
Analyzing the base distribution of sectors and the overall AI displacement clip.


```python
# Visualization 1: Displacement Ratio
plt.figure(figsize=(6,6))
df['displacement_occurred'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Displaced (1)', 'Retained (0)'], colors=['#E74C3C', '#2ECC71'], explode=(0.05, 0))
plt.title('Global Displacement Ratio Simulation')
plt.ylabel('')
plt.show()

```

```python
# Visualization 2: Sector Density vs Theoretical Exposure
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='theoretical_exposure_pct', y='sector', orient='h', order=df['sector'].value_counts().index)
plt.axvline(70, color='r', linestyle='--', label='Critical Risk Threshold (>70%)')
plt.title('Theoretical Automation Exposure by Sector')
plt.xlabel('Automation Exposure %')
plt.ylabel('Sector')
plt.legend()
plt.show()

```

## Task 2: Bivariate Analysis (Correlations)
### The Salary Paradox
Does a high salary protect you from displacement, or make you a prime target for efficiency deltas?


```python
# Visualization 3: Salary vs Displacement
df['salary_bin'] = pd.qcut(df['annual_salary_usd'], q=4, labels=['Q1 (Low)', 'Q2', 'Q3', 'Q4 (High)'])
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='salary_bin', y='displacement_occurred', palette='Blues_r')
plt.title('Displacement Rates across Salary Quartiles')
plt.ylabel('Average Displacement Rate')
plt.xlabel('Salary Bin')
plt.show()

```

### The Skill Shield
Mapping the impact of 'Educational Mismatch' on job security.


```python
# Visualization 4: The Skill Shield
df['gap_size'] = pd.qcut(df['education_mismatch_idx'], q=4, labels=['Low Gap', 'Mid-Low', 'Mid-High', 'High Gap'], duplicates='drop')
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='gap_size', y='displacement_occurred', marker='o', linewidth=2.5, color='#8E44AD')
plt.title('Displacement Trajectory vs Educational Mismatch (The Gap)')
plt.ylabel('Average Displacement Rate')
plt.xlabel('Competency Void / Gap Size')
plt.show()

```

## Task 3: Multivariate Analysis (Red Zones)
Isolating the extreme risk intersections: **Remote Eligibility + High Task Codifiability**.


```python
# Visualization 5: Heatmap of Remote vs Codifiability
df['high_codif'] = df['task_codifiability_score'] > 0.6
heat_df = df.groupby(['is_remote_eligible', 'high_codif'])['displacement_occurred'].mean().unstack()
plt.figure(figsize=(7, 5))
sns.heatmap(heat_df, annot=True, cmap='coolwarm', fmt='.1%')
plt.title('Heatmap: The Remote Danger Multiplier')
plt.ylabel('Remote Eligible (1=Yes)')
plt.xlabel('High Task Codifiability (>0.6)')
plt.show()

```

## Task 4: Segmenting the K-Shape Divergence
Contrasting the extreme resilience of Senior Strategic Orchestrators against the high vulnerability of Middle-Management Coordinators.


```python
# Segment Logic
df['worker_persona'] = 'Other'
df.loc[(df['is_senior'] == 1) & (df['annual_salary_usd'] > 120000), 'worker_persona'] = 'Strategic Orchestrator'
df.loc[(df['job_zone'].isin([3, 4])) & (df['is_senior'] == 0), 'worker_persona'] = 'Middle-Management Coordinator'

# Visualization 6: The K-Shape
k_shape_df = df[df['worker_persona'] != 'Other']
plt.figure(figsize=(8, 5))
sns.barplot(data=k_shape_df, x='worker_persona', y='displacement_occurred', palette=['#16A085', '#E67E22'])
plt.title('The K-Shaped Divergence in Automation Displacement')
plt.ylabel('Displacement Rate')
plt.xlabel('Worker Persona')
plt.show()

```

## Core Research Inferences

1. **The Remote Danger Multiplier**: Remote workers with highly codifiable tasks experience massive displacement rates compared to structurally protected on-site roles.
2. **The K-Shaped Divergence**: Middle-management is squeezed for efficiency, while high-income strategic workers leverage AI as an orchestrator.
3. **The Efficacy of the Skill Shield**: Specific dense integration skills (Cloud, Deep Learning) completely neutralize highly theoretical AI displacement exposures.
